```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd 
from pandas import DataFrame
import numpy as np
import seaborn as sns
import glob
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
folder = glob.glob('C:/Users/deniz/Documents/simul/shapes_2/*.txt')
data_list = []
for f in folder:
    print(f)
    data_list.append(f)
      
print(data_list)
```

```python
data_loaded=pd.DataFrame()
d=0
for i in data_list:
    datat = pd.read_csv(i, sep=" ", header=None)
    q=datat.iloc[1:202,1]
    d+=1
    datat=datat.iloc[1:202,3]
  
    
    #datat=datat[1:,[datat.iloc[1],datat.iloc[3]]
    
    #datat.drop(datat['0' , '2'] , axis='columns')
    #datat.drop(datat.index[[0,2,4,5]],axis=1)
    data_loaded = data_loaded.append(datat)
data_loaded.shape
transposed=data_loaded.T
print(transposed.shape)    
transposed.tail()
```

```python
col=np.arange(1, 91, 1).astype(str).tolist()
transposed.columns=col
transposed=transposed.astype(float)
transposed.head()
```

```python
df=transposed.set_index(q)

df=df.T
df.head()
```

```python
df['form-factor']=''
feat1=np.array(col).astype(int)
df = df.assign(number=feat1)

for index, row in df.iterrows():
    
    if row['number']>0 and row['number']<31:
        df.loc[index,'form-factor']='Sphere' 
    elif row['number']>50 and row['number']<71:
        df.loc[index,'form-factor']='Sphere' 
    else:
        df.loc[index,'form-factor']='Cylinder'
    
print(df.head())  
```

## Splitting data to the training and testing groups

```python
from sklearn.model_selection import train_test_split

# Extract the labels
labels = np.array(df.pop('form-factor'))

# 30% examples in test data
data_train, data_test, train_labels, test_labels = train_test_split(df, labels, 
                                                          stratify = labels,
                                                          test_size = 0.3, 
                                                          random_state = 0)
```

```python
data_train.shape
```

```python
data_test.shape
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gnb = GaussianNB()

pred = gnb.fit(data_train, train_labels).predict(data_test)

print("Naive-Bayes accuracy : ",accuracy_score(test_labels, pred, normalize = True))
```

# Linear SVC

```python
from sklearn.svm import LinearSVC

svc_model = LinearSVC(random_state=0,verbose=True,max_iter=1000)

pred = svc_model.fit(data_train, train_labels).predict(data_test)

print("LinearSVC accuracy : ",accuracy_score(test_labels, pred, normalize = True))
```

Increasing number of iterations didn't increase accuracy.

# K-neighbours

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=2)

neigh.fit(data_train, train_labels)

pred = neigh.predict(data_test)

print ("KNeighbors accuracy score : ",accuracy_score(test_labels, pred))
```

# Random Forest

```python
rndf = RandomForestClassifier(n_estimators=100, 
                               random_state=0, 
                               max_features = 'sqrt',
                               n_jobs=-1, verbose = 1)


rndf.fit(data_train, train_labels.T)
```

```python
n_nodes = []
max_depths = []

for ind_tree in rndf.estimators_:
    n_nodes.append(ind_tree.tree_.node_count)
    max_depths.append(ind_tree.tree_.max_depth)
    
print(f'Average number of nodes {int(np.mean(n_nodes))}')
print(f'Average maximum depth {int(np.mean(max_depths))}')
```

```python
train_rf_predictions = rndf.predict(data_train)
train_rf_probs = rndf.predict_proba(data_train)[:, 1]

rf_predictions = rndf.predict(data_test)
rf_probs = rndf.predict_proba(data_test)[:, 1]
```

```python
print ("Random forest accuracy score : ",accuracy_score(test_labels, rf_predictions))
```

I don't think that increasing the dataset will really help to increase the accuracy of these methods sufficiently. SAXS data is rather complicated since it gives integral characteristics of the system. In this case, only the impact from the form-factor was included. When structural factor will be added the complexity will increase. Adding anisotropy, polydispersity and noise will also make classification harder.
